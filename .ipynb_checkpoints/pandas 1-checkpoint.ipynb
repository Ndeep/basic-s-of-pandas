# pandas Replace method

```python
df_6=pd.read_csv("nyc_weather6.csv")
df_6
```

```python
#newdf_6=df_6.replace([-99999],np.NaN)#it will work complete dataframe
newdf_6=df_6.replace({
    'temperature':-99999,
    'windspeed':-99999,
    'event':'0'
},np.NaN)
newdf_6
```

```python
newdf_6=df_6.replace({
    -99999:np.NaN,
    'no event':'sunny'
},np.NaN,inplace=True)
newdf_6
```

```python
newdf_6=newdf_6.replace({
   'temperature':'[A-Za-z]',
    'windspeed':'[A-Za-z]'
},'',regex=True)
newdf_6
```

```python
import plotly
import plotly.graph_objs as go

# Create random data with numpy
import numpy as np

N = 100
random_x = new_df.index
random_y = new_df.Temperature
random_y1 = new_df.Humidity
random_y2 = new_df['Sea Level PressureIn']

# Create a trace
Temperature = go.Scatter(
    x = random_x,
    y = random_y,
    mode = 'markers',
    name="Temperature"
    
)
Humidity = go.Scatter(
    x = random_x,
    y = random_y1,
    mode = 'lines',
    name="Humidity"
)
Sea_Level_PressureIn = go.Scatter(
    x = random_x,
    y = random_y2,
    mode = 'lines+markers',
    name="Sea Level PressureIn"
)
data = [Temperature,Humidity,Sea_Level_PressureIn]

# Plot and embed in ipython notebook!
plotly.offline.plot(data, filename='basic-scatter.html')
```
